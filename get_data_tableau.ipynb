```python
import numpy as np
import pandas as pd
import requests
```

```python
url = "https://cea.nic.in/api/electricEnergySales.php"
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    data = response.json()
    df = pd.DataFrame(data)
    print(df.head())
else:
    print("Error: {response.status_code}")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.groupby('State').size().reset_index(name="Total Records")
```

```python
df['State'] = df['State'].replace({'[\.\$]':''}, regex=True)
```

```python
df.groupby('State').size().reset_index(name="Total Records")
```

```python
df['State'] = df['State'].str.strip()
```

```python
df.groupby('State').size().reset_index(name="Total Records")
```

```python
df['State'] = df['State'].replace('Madhay Pragesh','Madhya Pradesh')
```

```python
df.groupby('State').size().reset_index(name="Total Records")
```
